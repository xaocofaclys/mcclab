```python
import matplotlib.pyplot as plt 
import librosa.display
import numpy as np
y,sr = librosa.load(librosa.ex('trumpet'))
X = librosa.stft(y) 
X_db = librosa.amplitude_to_db(np.abs(X),ref=np.max)
fig,ax = plt.subplots()
img = librosa.display.specshow(X_db,ax=ax)
```
