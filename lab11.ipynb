```python
def Lzw(S):
    d = dict()
    for i in range(0,255):
        d[chr(i)] = i
    a = 1
    x = S[0]
    z = 256
    while(a != len(S)):
        nc = x+S[a]
        a+=1
        if nc in d.keys():
            x = nc
        else:
            d[nc] = z
            z+=1
            x = S[a-1]
            
    print(d)  
Lzw("TATAGATCTTAATATA")
```
