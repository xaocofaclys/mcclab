```python
import numpy as np 
from IPython.display import Audio

fs = 8000
duration = 2.0 
N = int(duration*fs)
f0 = 50
n = np.arange(N)
x = np.cos(2*np.pi*f0*n/fs)
Audio(x,rate=fs)
```

```python
import matplotlib.pyplot as plt 
import librosa.display
import numpy as np
y,sr = x,fs
X = librosa.stft(y) 
X_db = librosa.amplitude_to_db(np.abs(X),ref=np.max)
fig,ax = plt.subplots()
plt.title("frequency  50 hz ")
img = librosa.display.waveshow(X_db,ax=ax)
```
